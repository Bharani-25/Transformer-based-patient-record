```python

# Advanced transformer fine-tuning + evaluation: ROUGE, BLEU, METEOR + plots
# Run in Colab / Jupyter (with internet) — installs packages if missing.

import sys, subprocess, pkg_resources
import os

# Install required packages if missing (Colab/local with internet)
required = {"transformers", "rouge_score", "nltk", "torch", "scikit-learn", "pandas", "matplotlib"}
installed = {pkg.key for pkg in pkg_resources.working_set}
missing = required - installed
if missing:
    print("Installing missing packages:", missing)
    subprocess.check_call([sys.executable, "-m", "pip", "install", *missing])

# Imports
import random
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import T5Tokenizer, T5ForConditionalGeneration, get_linear_schedule_with_warmup
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

pd.set_option("display.precision", 3)

# -----------------------------
# 1) Create more realistic dummy dataset
# -----------------------------
random.seed(42)
np.random.seed(42)

conditions = [
    "hypertension", "diabetes", "asthma", "kidney stone", "fracture",
    "migraine", "bronchitis", "arthritis", "heart failure", "infection",
    "pneumonia", "stroke", "ulcer", "anemia", "thyroid disorder"
]

treatments = [
    "paracetamol", "antibiotics", "insulin", "nebulization",
    "painkillers", "aspirin", "hydration therapy", "physiotherapy",
    "antiviral drugs", "steroids", "anticoagulant", "surgery", "iron supplements"
]

templates = [
    "Patient presents with {cond} and reports symptoms for {days} days. Investigations show {finding}. Management initiated: {treat}. Advised follow-up in {followup} days.",
    "Admission due to {cond}. Clinical signs: {finding}. Treatment provided: {treat}. Discharge plan includes monitoring and follow-up.",
    "Outpatient visit for {cond}. Tests indicate {finding}. Started on {treat}. Return visit scheduled in {followup} days."
]

def gen_days():
    return np.random.randint(1, 15)

def gen_finding(cond):
    return f"moderate {cond} indicators" if cond not in ["fracture","kidney stone","stroke","ulcer"] else f"positive imaging for {cond}"

def generate_record():
    cond = random.choice(conditions)
    treat = random.choice(treatments)
    template = random.choice(templates)
    days = gen_days()
    followup = np.random.choice([3,5,7,14])
    finding = gen_finding(cond)
    return template.format(cond=cond, days=days, finding=finding, treat=treat, followup=followup)

def generate_summary_from_record(record):
    cond = next((c for c in conditions if c in record), "Condition")
    treat = next((t for t in treatments if t in record), "treatment")
    import re
    m = re.search(r"follow[- ]?up in (\d+)", record)
    follow = m.group(1) if m else "N/A"
    return f"{cond.capitalize()} managed with {treat}; follow-up in {follow} days."

N = 40
records = [generate_record() for _ in range(N)]
summaries = [generate_summary_from_record(r) for r in records]
df = pd.DataFrame({"record": records, "summary": summaries})

train_df, test_df = train_test_split(df, test_size=0.25, random_state=42)
print(f"Dataset: {len(train_df)} train samples, {len(test_df)} test samples")

# -----------------------------
# 2) Tokenizer, model and dataset classes
# -----------------------------
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

class EHRDataset(Dataset):
    def __init__(self, df, tokenizer, max_len=256, target_max_len=64):
        self.df = df.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.target_max_len = target_max_len

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        record = self.df.loc[idx, "record"]
        summary = self.df.loc[idx, "summary"]

        input_enc = self.tokenizer(
            "summarize: " + record,
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )

        target_enc = self.tokenizer(
            summary,
            truncation=True,
            padding="max_length",
            max_length=self.target_max_len,
            return_tensors="pt"
        )

        labels = target_enc["input_ids"]
        labels[labels == tokenizer.pad_token_id] = -100

        return {
            "input_ids": input_enc["input_ids"].squeeze(0),
            "attention_mask": input_enc["attention_mask"].squeeze(0),
            "labels": labels.squeeze(0)
        }

train_dataset = EHRDataset(train_df, tokenizer)
test_dataset = EHRDataset(test_df, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# -----------------------------
# 3) Training setup
# -----------------------------
EPOCHS = 3
optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4)
num_training_steps = EPOCHS * len(train_loader)
scheduler = get_linear_schedule_with_warmup(
    optimizer, num_warmup_steps=max(1, int(0.1 * num_training_steps)), num_training_steps=num_training_steps
)

print(f"Training for {EPOCHS} epochs, {num_training_steps} steps (approx). Device: {device}")

# -----------------------------
# 4) Training loop
# -----------------------------
model.train()
for epoch in range(EPOCHS):
    total_loss = 0.0
    for batch in train_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{EPOCHS} — avg loss: {avg_loss:.4f}")

# -----------------------------
# 5) Inference & evaluation
# -----------------------------
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
smooth_fn = SmoothingFunction().method4

results = []
model.eval()
with torch.no_grad():
    for i, row in test_df.reset_index(drop=True).iterrows():
        rec = row["record"]
        ref = row["summary"]

        inputs = tokenizer("summarize: " + rec, return_tensors="pt", truncation=True, max_length=256).to(device)
        gen_ids = model.generate(
            inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_length=64,
            num_beams=6,
            early_stopping=True,
            no_repeat_ngram_size=2,
            length_penalty=1.0
        )
        pred = tokenizer.decode(gen_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)

        rouge_scores = scorer.score(ref, pred)
        ref_tokens = nltk.word_tokenize(ref.lower())
        pred_tokens = nltk.word_tokenize(pred.lower())
        try:
            bleu = sentence_bleu([ref_tokens], pred_tokens, smoothing_function=smooth_fn)
        except Exception:
            bleu = 0.0
        try:
            meteor = meteor_score([ref], pred)
        except Exception:
            meteor = 0.0

        results.append({
            "record": rec,
            "reference": ref,
            "prediction": pred,
            "rouge1": rouge_scores['rouge1'].fmeasure,
            "rouge2": rouge_scores['rouge2'].fmeasure,
            "rougeL": rouge_scores['rougeL'].fmeasure,
            "bleu": bleu,
            "meteor": meteor
        })

results_df = pd.DataFrame(results)
avg_metrics = results_df[["rouge1","rouge2","rougeL","bleu","meteor"]].mean().to_dict()

print("\nAverage evaluation metrics on test set:")
for k,v in avg_metrics.items():
    print(f"  {k}: {v:.3f}")

# Display results dataframe
display(results_df)

# -----------------------------
# 6) Plotting (matplotlib only; one chart each)
# -----------------------------
plt.figure(figsize=(8,4))
plt.bar(range(len(results_df)), results_df["rouge1"])
plt.xlabel("Test sample index")
plt.ylabel("ROUGE-1 F1")
plt.title("ROUGE-1 per test sample")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
plt.bar(range(len(results_df)), results_df["rougeL"])
plt.xlabel("Test sample index")
plt.ylabel("ROUGE-L F1")
plt.title("ROUGE-L per test sample")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
plt.bar(range(len(results_df)), results_df["bleu"])
plt.xlabel("Test sample index")
plt.ylabel("BLEU (sentence)")
plt.title("BLEU per test sample")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
plt.bar(range(len(results_df)), results_df["meteor"] * 100)
plt.ylim(0, 100)
plt.grid(axis="y")

plt.xlabel("Test sample index")
plt.ylabel("METEOR")
plt.title("METEOR per test sample")
plt.tight_layout()
plt.show()

plt.figure(figsize=(7,4))
avg_vals = [avg_metrics["rouge1"], avg_metrics["rouge2"], avg_metrics["rougeL"], avg_metrics["bleu"], avg_metrics["meteor"]]
labels = ["ROUGE-1","ROUGE-2","ROUGE-L","BLEU","METEOR"]
plt.bar(labels, avg_vals)
plt.ylabel("Score (0-1)")
plt.title("Average evaluation metrics on test set")
plt.tight_layout()
plt.show()

# Save results
os.makedirs('/mnt/data', exist_ok=True)
results_df.to_csv("/mnt/data/eval_results.csv", index=False)
print("\nSaved detailed results to /mnt/data/eval_results.csv")
```
